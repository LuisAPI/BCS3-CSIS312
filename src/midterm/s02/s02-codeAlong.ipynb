```python
import numpy as np

#define the transition matrix P. States: sunny, rainy, cloudy
P = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.4, 0.3],
    [0.4, 0.3, 0.3]
])

#define the states
states = ['Sunny', 'Rainy', 'Cloudy']

#create a function to predict the weather
def predict_weather(current_state, P):
    if current_state not in states:
      raise ValueError(f"Invalid state '{current_state}'.")

    state_index = states.index(current_state)

    probabilities = P[state_index]

    next_state_index = np.random.choice(len(states), p = probabilities)
    return states[next_state_index]

def get_user_input():
    while True:
        current_weather = input(f"Enter the current weather state ({', '.join(states)}): ").capitalize()
        if current_weather in states:
            return current_weather
        else:
            print("Invalid input. Please enter one of the following: Sunny, Rainy and Cloudy.")

#main program
current_weather = get_user_input()
next_weather = predict_weather(current_weather, P)


print(f'Current weather : {current_weather}')
print(f"Predicted weather for tomorrow: {next_weather}")
```

```python
import numpy as np

#probability of having the disease (prior probability)
P_A = 0.01
#probability of testing positive given the disease (sensitivity)
P_B_given_A = 0.95
#probability of testing positive without the disease (false positive rate)
P_B_given_not_A = 0.05
#probability of not having the disease
P_not_A = 1 - P_A

#calculate the marginal likelihood P(B)
P_B = (P_B_given_A * P_A) + (P_B_given_not_A * P_not_A)

#apply bayes theorem to find P(A|B)
P_A_given_B = (P_B_given_A * P_A) / P_B

print('P(B)', P_B)
print('P(A|B)', P_A_given_B)

#interpretation
#P(B) the probability of testing positive is approx 5.9%
#P(A|B) given the positive result, the probability of actually having the disease is approx 16%
```
