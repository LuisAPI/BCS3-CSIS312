Levels of languages
- phonology
  - the study of the sound system of a language. Phonological analysis is crucial tasks like speech recognition and synthesis
    - understanding the sound patterns is essential for converting spoken language into text
  - phonemes
    - are the smallest units of sound in a language
- morphology
  - the study of the structure of words, it involves understanding how words are formed from morphemes
    - smallest units of meaning
  - morphological analysis helps in tasks like stemming, where words are reduced to their base form
    - running to run
    - walking to walk
    - eating to eat
- syntax
  - the study of the structure of sentences
    - syntax rules determine how words are arranged in form grammatically form
    - syntactic analysis is used for parsing sentences (machine translation)
- semantics
  - the study of meaning in language
    - semantic analysis focuses on understanding the meaning of words, phrases and sentences
- pragmatics
  - the study of how context influences the interpretation of meaning
    - pragmatic analysis considers the speaker's intent, the relationship between the speaker and listener and other contextual factors

N-grams
  - continuous sequence of n item and from a given sample of text
  - (NLP) is used to predict what next word in a sequence by looking at the previous (n-1) word
    - bi-gram(2-gram model) is used to predict the next word based on the previous word "I love you"

```python
import random
from collections import defaultdict

'''
# default value 0
#defaultdict subclass for Python's built-in dict(dictionary) that returns a default
#if the requested key does not exist
d = defaultdict(int)
d['apply'] += 1
d['remove'] += 2
d['update'] += 3

print(d)
'''

#sample text for n-gram
text = "I love Pia. I like Nicole."

#create a bigram model
n = 2
ngrams = defaultdict(list)
words = text.split()

for i in range(len(words) - n + 1):
  gram = tuple(words[i:i+n])
  next_word = words[i + n] if i + n < len(words) else None
  #add next word if it's not None
  if next_word:
      ngrams[gram].append(next_word)

#user input
user_input = input("Enter a word or phrase: ")
inpt_words = user_input.split()

"""
#generate random text using bigram model
#choose a gram that has a next words
current_gram = random.choice([gram for gram in ngrams if ngrams[gram]])
result = list(current_gram)

#generate 10 words
for _ in range(10):
  if current_gram in ngrams and ngrams[current_gram]:
    next_word = random.choice(ngrams[current_gram])
    result.append(next_word)
    current_gram = tuple(result[-n:])
  else:
    break

generated_text = ' '.join(result)
print(generated_text)
"""

# find matching grams
matching_grams = [gram for gram in ngrams if gram[-1] == inpt_words[-1]]

if matching_grams:
  current_gram = random.choice(matching_grams)
  result = list(current_gram)

  #autocomplete
  for _ in range(10):
    if current_gram in ngrams and ngrams[current_gram]:
      next_word = random.choice(ngrams[current_gram])
      result.append(next_word)
      current_gram = tuple(result[-n:])
    else:
      break

  print("Autocomplete suggestion : ", ' '.join(result))
else:
  print("No matching grams found.")
```
